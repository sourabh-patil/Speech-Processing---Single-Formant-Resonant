# Single Formant Resonator filter:magnitude response and impulse response

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import freqz
```

```python
f_range = np.linspace(0,8000,1000)     ### Creating a linearspace of 8000 points between 0 to 1000 as we want to 
                                       ### plot the graph in Hz. 8000 is nothing but fs/2 to represent it as pi
                                       ### which will be spit out by 'freqz' function
```

```python
w , h = freqz([1],[1,-1.804,0.9245],worN= 1000)    ### Calculating w means normalized frequencies and h means
                                                   ### frquency response as complex numbers. Frequncy response
                                                   ### is computed at 1000 frequencies as denoted in freqz
                                                   ### Calculated the values of numerator and denominator using
                                                   ### given parameters 
```

```python
plt.figure( figsize = (15,5) )
plt.title('Magnitude response of Single Formant Resonator')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid()
plt.plot(f_range, 20 * np.log10(abs(h)))
```

```python
#### First defining a function to generate impulse 
def d(n):                                
    if(n ==0):    
        d = 1
    else:
        d = 0
    return d
#### Implementing calculated difference equation using for loop. I am plotting the time domain waveform for 200 samples for time being
h = np.zeros(202)
h[0] = 0
h[1] = 0
for i in range(200):
    h[i+2] = d(i) + 1.804*h[i+1] - 0.9245*h[i]
```

```python
import matplotlib.pyplot as plt
n = np.arange(200)
plt.figure(figsize=(15,5))
plt.title('Impulse response of Single Formant Resonator in time domain')
plt.xlabel('Number of samples')
plt.ylabel('Amplitude levels')
plt.stem(n, h[2:203])
```
